```python
import pandas as pd
data = pd.read_csv("/content/breast_cancer_survival[1].csv")
data
```

```python
data.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
```

```python
data_cleaned = data.dropna()

numeric_data = data.select_dtypes(include=['float64', 'int64'])
data_filled = data.fillna(numeric_data.mean())
```

```python
numeric_columns = ['Age', 'Protein1', 'Protein2', 'Protein3', 'Protein4']
data[numeric_columns] = data[numeric_columns].apply(pd.to_numeric, errors='coerce')
```

```python
features = data_cleaned.drop(columns=['Patient_Status'])
target = data_cleaned['Patient_Status']
features_encoded = pd.get_dummies(features)

accuracies = []
```

```python
for test_size in [0.2, 0.3, 0.4]:
    X_train, X_test, y_train, y_test = train_test_split(features_encoded, target, test_size=test_size, random_state=42)

    model = RandomForestClassifier(random_state=42)
    model.fit(X_train, y_train)

    accuracy = model.score(X_test, y_test)
    accuracies.append((test_size, accuracy))

for size, acc in accuracies:
    print(f'Test Size: {size}, Accuracy: {acc:.2f}')
```

Implement Logistic Regression using pre-defied library. Analysis of different training and testing splits ranges, and find the accuracy.

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)
```

```python
accuracies = []

test_sizes = [0.2, 0.3, 0.4]

for test_size in test_sizes:
    X_train, X_test, y_train, y_test = train_test_split(features_scaled, target, test_size=test_size, random_state=42)

    model = LogisticRegression(max_iter=2000)
    model.fit(X_train, y_train)

    predictions = model.predict(X_test)
    accuracy = accuracy_score(y_test, predictions)
    accuracies.append((test_size, accuracy))

for size, acc in accuracies:
    print(f'Test Size: {size}, Accuracy: {acc:.2f}')
```
